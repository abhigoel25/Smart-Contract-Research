# aTypes Tutorial

Agentics overloads `&` operator for merging two aTypes as well as instances,.

```python
from agentics.core.transducible_functions import  Transduce, With, transducible
from pydantic import BaseModel
from typing import Optional
from agentics.core.atype import *

class Email(BaseModel):
    subject:Optional[str]
    name:Optional[str]

class User(BaseModel):
    name:Optional[str] =None
    last_name:Optional[str] = None
    full_name:Optional[str] = None

email = Email(subject="Hello", name="Alfio")
user = User(name="Isa")
```

### Merging Instances

```python

#Meriging of attributes with different values prefer right operand 
merged_instance1 = email & user
print (merged_instance1)
merged_instance2 = user & email
print (merged_instance2)
```

### Merging types

```python
MergedType= Email & User
print(MergedType)
print(isinstance(merged_instance1,MergedType ))
print(isinstance(merged_instance2,MergedType ))
```

the Merged types is subclass of both types that generated it

```python
print(issubclass(User,MergedType ))
print(issubclass(Email,MergedType ))
print(issubclass(MergedType, User ))
print(issubclass(MergedType, Email ))
```

Types can be semantically merged , returning an instance which is the average of the two inputs

```python
from pydantic import BaseModel
from agentics.core.transducible_functions import semantic_merge

print(await semantic_merge(User(name="Alfio"), User(last_name="Gliozzo")))
```
