# aTypes Tutorial

Agentics overloads `&` operator for merging two aTypes as well as instances,.

```python
from pydantic import BaseModel, Field
from typing import Optional
import agentics.core.atype
from pprint import pprint
from agentics import AG

class Movie(BaseModel):
    movie_name: Optional[str] = None
    description: Optional[str] = None
    year: Optional[int] = None

class Genre(BaseModel):
    genre:Optional[str] = Field(
        None,
        description="return one category,   "
                    "e.g. comedy, drama, ...")
  

movies = AG.from_csv(
    "data/movies.csv", atype=Movie, max_rows=2
) 
```

#### Type Composition

```python
ComposedType = Genre @ Movie
print(ComposedType.model_fields)
```


{'source': FieldInfo(annotation=Movie),


 'target': FieldInfo(annotation=Genre)}

### Type Merge

```python
MergedType = Genre & Movie
pprint(MergedType.model_fields)
```

### Merging types

the Merged types is subclass of both types that generated it

```python
print(issubclass(Genre,MergedType ))
print(issubclass(Movie,MergedType ))
print(issubclass(MergedType, Genre ))
print(issubclass(MergedType, Movie ))
```

```python
print(movies[0] & movies[1] == movies[1])
print(movies[1] & movies[0] == movies[1])
movie=movies[1]
```

Types can be semantically merged , returning an instance which is the average of the two inputs

#### State Composition

```python
comp_inst = movie @ Genre(genre="Crime")
comp_inst.model_dump()
```
