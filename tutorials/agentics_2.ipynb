```python
from pydantic import BaseModel
from agentics.core.agentics_2 import AgenticsTransduction

class Answer(BaseModel):
    answer: str


class Question(BaseModel):
    question: str


async def main():
    agent = AgenticsTransduction(atype=Answer, provide_explanations=True)
    response = await (
        agent
        << AgenticsTransduction(
            atype=Question, states=[Question(question="What is the capital of France?")]
        )
    )
    print(response.explanations)

await main()
```
